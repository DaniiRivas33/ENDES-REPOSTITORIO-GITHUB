```python
from datetime import datetime
class Alumno:
    def __init__(self, nombre, apellido1, apellido2, fecha_nacimiento):
        self.nombre = nombre
        self.apellido1 = apellido1
        self.apellido2 = apellido2
        self.fecha_nacimiento = fecha_nacimiento  
        self.notas = {}  

    def atributos(self):
        print(f"Nombre completo: {self.nombre} {self.apellido1} {self.apellido2}")
        mayor = "Sí" if self.mayor_edad() else "No"
        print(f"¿Mayor de edad?: {mayor}")

    def mayor_edad(self):
        fecha_nac = datetime.strptime(self.fecha_nacimiento, '%Y-%m-%d')
        edad = (datetime.now() - fecha_nac).days // 365
        return edad >= 18

    def imprimir_y_actualizar_notas(self):
        self.atributos()
        print("\nNotas actuales:")
        
        if not self.notas:
            print("No hay notas registradas.")
        else:
            
            for asignatura, nota in self.notas.items():
                print(f"{asignatura}: {nota}")

            
            for asignatura in self.notas:
                cambiar = input(f"¿Quieres cambiar la nota de {asignatura}? (si/no): ").strip().lower()
                if cambiar == 'si':
                    nueva_nota = int(input(f"Introduce la nueva nota (entero) para {asignatura}: "))
                    self.notas[asignatura] = nueva_nota  

    def poner_notas(self, asignatura, nota):
        self.notas[asignatura] = nota



alumno = Alumno(nombre="Alejandro", apellido1="Beguer", apellido2="Ruiz", fecha_nacimiento="2006-03-10")


asignaturas_notas = {
    "Programación": 1,
    "Base de Datos": 9,
    "Lenguaje de Marcas": 3,
    "IP": 5,
    "Inglés": 8,  
    "ENDES": 9     
}


for asignatura, nota in asignaturas_notas.items():
    alumno.poner_notas(asignatura, nota)


alumno.imprimir_y_actualizar_notas()


print("\nNotas finales actualizadas:")
for asignatura, nota in alumno.notas.items():
    print(f"{asignatura}: {nota}")

```
